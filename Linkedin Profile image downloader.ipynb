```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.firefox.options import Options
import time
import pandas as pd
from PIL import Image
im = Image.open("./2022 images/dummy.png")#Loading dummy image
csv=pd.read_csv('.\Major Project-2022.csv')#File containing linkedin id urls
df=pd.DataFrame({'Roll':csv['Roll'],'Linkedin':csv['Linkedin']})
#df=df.dropna()

def document_initialised(driver):
    return driver.execute_script("return initialised")

options = Options()
options.add_argument("--headless")
driver = webdriver.Firefox(options=options)
driver.get("https://www.linkedin.com/login")
username = driver.find_element(By.ID, "username")
username.clear()
username.send_keys("")#Enter your email address

password = driver.find_element(By.ID,"password")
password.clear()
password.send_keys("")#Enter your linkedin password

# subimit = driver.find_element(By.NAME, "Sign in")
password.submit()

time.sleep(13)
for i in range(1,len(df['Roll'])):
    try:
        print(df.iloc[i]['Linkedin'],df.iloc[i]['Roll'])#For showing current progress
        driver.switch_to.new_window('tab')
        driver.get(df.iloc[i]['Linkedin'])#Getting linkedin profile url
        element = driver.find_element(By.CLASS_NAME,"pv-top-card-profile-picture__image")#Selecting Profile Image Class
        src = element.get_attribute("src")
        driver.switch_to.new_window()#Opening image in new window
        driver.get(src)
        driver.save_screenshot('./2022 images/'+df.iloc[i]['Roll']+'.png')#Downloading image
    except:
        try:
            im=im.save("./2022 images/"+df.iloc[i]['Roll']+'.png')#If link is invalid we will create a dummy image
        except:
            continue




# driver.navigate('https://www.linkedin.com/feed')
# WebDriverWait(driver).until(document_initialised)



```
